```python
import pandas as pd
import numpy as np

# Sistema especialista para classificar Cadernos
class Especialist:
    # Construtor
    def __init__(self, altura, largura, possui_mola, espessura):
        self.altura = altura
        self.largura = largura
        self.possui_mola = possui_mola
        self.espessura = espessura

    # Classificador
    def classify(self):

        # Regras para cadernos Universitários
        if 27 <= self.altura <= 28 and 19 <= self.largura <= 21:
            if 5 <= self.espessura < 8:
                return "Caderno com 1 matéria"
            elif 8 <= self.espessura < 12:
                return "Caderno com 10 matérias"
            elif 12 <= self.espessura < 15:
                return "Caderno com 12 matérias"
            elif self.espessura >= 15:
                return "Caderno com 15 matérias"

        # Regras para cadernos Pequeno
        elif 12 <= self.altura <= 13 and 9 <= self.largura <= 10:
            if 5 <= self.espessura < 8:
                return "Mini caderno de anotações" if self.possui_mola == 1 else "Caderno de bolso"
            elif 12 <= self.espessura < 15:
                return "Agenda" if self.possui_mola == 1 else "Diário"
            else:
                return "Caderno pequeno"

        # Regras para cadernos A5
        elif 21 <= self.altura <= 22 and 15 <= self.largura <= 16:
            if 5 <= self.espessura < 8:
                return "Caderninho" if self.possui_mola == 1 else "Caderno de anotações"
            elif 12 <= self.espessura < 15:
                return "Fichário" if self.possui_mola == 1 else "Diário grande"
            else:
                return "Caderno A5"

        return "Caderno desconhecido"

    # Inicia e classifica os objetos vindos da Database
    def classify_from_Cadernos(Cadernos):
        classifications = []

        for _, row in Cadernos.iterrows():
            especialista = Especialist(
                altura=row['Altura (cm)'],
                largura=row['Largura (cm)'],
                possui_mola=row['Possui Mola'],
                espessura=row['Espessura (mm)']
            )

            classifications.append(especialista.classify())

        Cadernos['Classificação'] = classifications
        return Cadernos

# Lê os arquivos da Database, necessário "incluir" o cadernos.csv (Trabalho 2) ali na esquerda na pasta arquivos
Cadernos = pd.read_csv('cadernos.csv')
Classificado = Especialist.classify_from_Cadernos(Cadernos)
print(Classificado[['Altura (cm)', 'Largura (cm)', 'Possui Mola', 'Espessura (mm)', 'Classificação']])
```
